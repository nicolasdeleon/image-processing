```
img = cv2.imread('palabras-defo.png', cv2.IMREAD_GRAYSCALE)

img_contour = cv2.imread('contour.png', cv2.IMREAD_GRAYSCALE)

def plot_images(original, txt1, other, txt2):
    fig=plt.figure(figsize=(14, 10))
    fig.add_subplot(1, 2, 1)
    plt.title(txt1)
    plt.imshow(original, cmap='gray')
    fig.add_subplot(1, 2, 2)
    plt.title(txt2)
    plt.imshow(other, cmap='gray')
```

## Kernel and structural elements

```
kernel_cross_3 =  cv2.getStructuringElement(cv2.MORPH_CROSS, (3, 3))
print(kernel_cross_3)

kernel_ones_3 = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))

"""
This structural element comes pre-defined in cv2 as cv2.MORPH_CROSS
B = [
     [0, 1, 0],
     [1, 1, 1],
     [0, 1, 0]
]
""";
```

## Ejercicio 1

```
img_dilation = cv2.dilate(img, kernel_cross_3, iterations=2)
plot_images(img, 'original', img_dilation, 'dilatada')
```

## Ejercicio 2

```
img_contour_erosion = cv2.erode(img_contour, kernel_cross_3, iterations=1)
plot_images(img_contour, 'original', img_contour - img_contour_erosion, 'contour by erosion')
```
