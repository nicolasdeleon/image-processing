```
def plot_images(original,txt1,degraded,txt2):
    fig=plt.figure(figsize=(14, 10))
    fig.add_subplot(1, 2, 1)
    plt.title(txt1)
    plt.imshow(original,cmap='gray')
    fig.add_subplot(1, 2, 2)
    plt.title(txt2)
    plt.imshow(degraded,cmap='gray')
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('stars.PNG',cv2.IMREAD_GRAYSCALE)
kernel =  cv2.getStructuringElement(cv2.MORPH_CROSS,(2,2)) #np.array([[1,1,1],[1,1,1],[1,1,1]]) #np.ones((5,5),np.uint8)
print(kernel)

kernel = np.ones((25,25),np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

#dilation = cv2.dilation(img,kernel,iterations = 2)
#erosion = cv2.erosion(img,kernel,iterations = 2)
plot_images(img,"original",opening,"opening")
```
